In this notebook we will see some of the univariate charts developed using *seaborn* library.

# Count Plot

```python
#importing library
import seaborn as sns
import matplotlib.pyplot as plt

#setting theme
sns.set_theme(style="darkgrid")

#importing dataset
titanic = sns.load_dataset("titanic")

#plotting chart
ax = sns.countplot(x="class", data=titanic, hue="sex", palette="Blues")
plt.show()
```

# Histogram

```python
#importing library
import seaborn as sns

#setting theme
sns.set_theme(style="darkgrid")

#importing dataset
titanic = sns.load_dataset("titanic")

#plotting chart
sns.histplot(data=titanic, x="age", kde=True, bins = 5, hue = 'sex')
plt.show()
```

# Box Plot

```python
#importing library
import seaborn as sns

#setting theme
sns.set_theme(style="darkgrid")

#importing dataset
titanic = sns.load_dataset("titanic")

#plotting chart
sns.boxplot(data=titanic, x="class", y='age', hue ='sex', palette="Greens")
plt.show()
```

```python
#importing library
import seaborn as sns

#setting theme
sns.set_theme(style="darkgrid")

#importing dataset
titanic = sns.load_dataset("titanic")

#plotting chart
sns.scatterplot(data=titanic, x="age", y='fare', hue ='alive', palette="Greens")
plt.show()
```

# Joint Plot

```python
#importing library
import seaborn as sns

#setting theme
sns.set_theme(style="darkgrid")

#importing dataset
titanic = sns.load_dataset("titanic")

#plotting chart
sns.jointplot(data=titanic, x="age", y='fare', hue ='alive', palette="Greens")
plt.show()
```

# Joint Plot with Regression Line

```python
#importing library
import seaborn as sns

#setting theme
sns.set_theme(style="darkgrid")

#importing dataset
titanic = sns.load_dataset("titanic")

#plotting chart
sns.jointplot(data=titanic, x="age", y='fare', palette="Greens", kind="reg")
plt.show()
```

# Pair Plot

```python
#importing library
import seaborn as sns

#setting theme
sns.set_theme(style="darkgrid")

#importing dataset
titanic = sns.load_dataset("titanic")

#plotting chart
sns.pairplot(titanic, hue="alive")
plt.show()
```
