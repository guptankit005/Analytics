```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
```

```python
data = pd.read_csv(r'C:\Users\ankit.bo.kumar\dataset\titanic\train.csv')
data.head()
```

```python
#data.isnull(data['Age'][17])
for i in range(len(data['Age'])):
    if pd.isnull(data['Age'].iloc[i])==True:
        if data['Pclass'].iloc[i]=='3':
            data['Age'].iloc[i]='24'
        elif data['Pclass'].iloc[i]=='2':
            data['Age'].iloc[i]='29'
        else:
            data['Age'].iloc[i]='37'
```

```python
data.drop('Cabin',axis=1,inplace=True)
data.dropna(inplace=True)
data.head()
```

```python
data.drop('Fare',axis=1,inplace=True)
data.dropna(inplace=True)
data.head()
```

```python
sex = pd.get_dummies(data['Sex'])
embark = pd.get_dummies(data['Embarked'])
#drop the sex,embarked,name and tickets columns
data.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
#concatenate new sex and embark column to our train dataframe
data = pd.concat([data,sex,embark],axis=1)
```

```python
#check the head of dataframe
data.head(10)
```

```python
X_train = data[['PassengerId','Pclass','Age','SibSp','Parch','female','male','C','Q','S']]
X_train.head()
```

```python
y_train = data[['Survived']]
y_train.head()
```

```python
nB = GaussianNB()
```

```python
nB.fit(X_train,y_train)
```

```python
test_data = pd.read_csv(r'C:\Users\ankit.bo.kumar\dataset\titanic\test.csv')
test_data.head()
```

```python
for i in range(len(test_data['Age'])):
    if pd.isnull(test_data['Age'].iloc[i])==True:
        if test_data['Pclass'].iloc[i]=='3':
            test_data['Age'].iloc[i]='25'
        elif test_data['Pclass'].iloc[i]=='2':
            test_data['Age'].iloc[i]='27'
        else:
            test_data['Age'].iloc[i]='42'
```

```python
sex = pd.get_dummies(test_data['Sex'])
embark = pd.get_dummies(test_data['Embarked'])
#drop the sex,embarked,name and tickets columns
test_data.drop(['Sex','Embarked','Name','Ticket','Fare'],axis=1,inplace=True)
#concatenate new sex and embark column to our train dataframe
test_data = pd.concat([test_data,sex,embark],axis=1)
```

```python
X_test = test_data[['PassengerId','Pclass','Age','SibSp','Parch','female','male','C','Q','S']]
X_test.head()
X_test.info()
```

```python
#predictions
y_pred = nB.predict(X_test)
```
