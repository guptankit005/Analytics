```python
import openpyxl 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
xl_doc = openpyxl.load_workbook('DataMatching.xlsx')
print (type(xl_doc))
```

```python
#to check what sheets we have in our Excel document
xl_doc.get_sheet_names()
```

```python
# to access column A row 2 in the Excel document
sheet = xl_doc.get_sheet_by_name('Sheet1')
print(sheet['A2'].value)
```

```python
#same thing what we did in above line of code can also be done by stating the numeric value for row and column
sheet.cell(row=2,column=1).value
```

```python
""""this block of code first matches the country from first column and then from third column, and if it exists
in the third column then the corresponding value from forth column for the country is copied and pasted in fifth
column correspondingly to the row belonging for the country in first column"""
for i in range(1,152):
    for j in range(1,192):
        if sheet.cell(row=i,column=1).value == sheet.cell(row=j, column=3).value:
            sheet.cell(row=i,column=5).value = sheet.cell(row=j,column=4).value
            xl_doc.save(r"C:\Users\ankit.bo.kumar\DataMatching_Raw.xlsx")      
            
   
                                  
```
